```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SpotifyFeatures.csv')
```

**Information about dataset**

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

**Correlation between each column**

```python
corr = df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr,annot=True)
```

**Visualization**

Most popular songs are dancable and easy to sing, hence, logically they should have high danceability and low speechiness.
Furthermore, their duration is usually less than 5 minutes.

```python
sns.scatterplot(x=df.duration_ms, y=df.popularity)
```

```python
sns.scatterplot(x=df.danceability, y=df.popularity)
```

```python
sns.scatterplot(x=df.speechiness, y=df.popularity)
```

**Popularity by genre, key, and time signature**

```python
plt.figure(figsize=(30,8))
sns.countplot(df.genre)
```

```python
sns.catplot(x = "popularity", y = "genre", kind = "bar" ,
            palette = "pastel", edgecolor = ".6",
            data = df)
```

```python
sns.catplot(x = "popularity", y = "key", kind = "bar" ,
            palette = "pastel", edgecolor = ".6",
            data = df)
```

```python
sns.catplot(x = "popularity", y = "time_signature", kind = "bar" ,
            palette = "pastel", edgecolor = ".6",
            data = df)
```
